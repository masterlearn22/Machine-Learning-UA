```python

import pandas as pd

# ‚úÖ Fungsi kategorisasi otomatis tanpa menambah suffix
def kategorisasi_otomatis(df, kolom, bins, labels):
    if len(bins) - 1 != len(labels):
        raise ValueError(f"‚ùå Jumlah bins dan labels tidak sesuai pada kolom '{kolom}'")
    if len(set(labels)) != len(labels):
        raise ValueError(f"‚ùå Label duplikat terdeteksi pada kolom '{kolom}': {labels}")
    # Menimpa kolom lama langsung
    df[kolom] = pd.cut(df[kolom], bins=bins, labels=labels, include_lowest=True, ordered=True)
    return df
```

```python

# ‚úÖ Definisi kategori (pastikan nama kolom sesuai dataset)
kolom_kategori = {
    'Hours_Studied': ([0, 10, 20, 30], ['Low', 'Medium', 'High']),
    'Attendance': ([0, 74, 100], ['Not Met', 'Met']),
    'Sleep_Hours': ([0, 4, 6, 8, 10], ['Low', 'Medium', 'High', 'Extreme']),
    'Previous_Scores': ([0, 40, 54, 62, 70, 78, 86, 100], ['E', 'D', 'C', 'BC', 'B', 'AB', 'A']),
    'Tutoring_Sessions': ([0, 2, 4, 6, 8], ['Low', 'Medium', 'High', 'Extreme']),
    'Physical_Activity': ([0, 2, 4, 6], ['Low', 'Medium', 'High']),
    'Exam_Score': ([0, 40, 54, 62, 70, 78, 86, 100], ['E', 'D', 'C', 'BC', 'B', 'AB', 'A'])
}
```

```python

# ‚úÖ Baca dataset (ubah nama file sesuai kebutuhan)
df = pd.read_excel("StudentPerformanceFactors.xlsx")

# Simpan urutan kolom asli
kolom_awal = df.columns.tolist()

print("Kolom di dataset awal:")
print(kolom_awal)
```

```python

# ‚úÖ Terapkan kategorisasi otomatis tanpa hapus kolom
for kolom, (bins, labels) in kolom_kategori.items():
    if kolom in df.columns:
        df = kategorisasi_otomatis(df, kolom, bins, labels)
        print(f"‚úîÔ∏è Kolom '{kolom}' berhasil dikategorikan ulang")
```

```python

# ‚úÖ Urutkan kolom kembali ke posisi awal
kolom_final = [kol for kol in kolom_awal if kol in df.columns]
df = df[kolom_final]

print("üìã Urutan kolom telah dikembalikan seperti semula.")
```

```python

# ‚úÖ Simpan hasil akhir ke Excel
df.to_excel("hasil_dataset_kategorisasi_final_v2.xlsx", index=False)
print("üìÅ Dataset akhir telah disimpan ke 'hasil_dataset_kategorisasi_final_v2.xlsx'")
df.head()
```
